```python
# Importar las librerías necesarias
import pandas as pd
```

```python
# Cargar el archivo CSV
data = pd.read_csv("Patient_info_and_summary.csv")
```

```python
# Mostrar los 5 primeros datos
data.head(5)
```

```python
# Información del dataset
data.info()
```

```python
# Crear la columna 'General patient information 1'
data['General patient information'] = (
    "Name of patient: " + data['Name'] + 
    ", Patient's date of birth: " + data['Date_of_Birth'] + 
    ", Patient address: " + data['Address'] + 
    ", Patient Progress Summary: " + data['Patient Progress Summary']
)

# Crear la columna 'Patient_ID' enumerada desde 1 hasta el número de filas
data['Patient_ID'] = range(1, len(data) + 1)
```

```python
# Mostrar los 5 primeros datos
data.head(5)
```

```python
# Exportar el DataFrame modificado a un nuevo archivo CSV
data.to_csv("Patient_info_and_summary_updated.csv", index=False)

print("El archivo CSV ha sido exportado exitosamente.")
```
